# 분석메모

평가지표: Accuracy

- id : 식별 고유값  

- fixed acidity : 고정(비휘발성) 산도: 와인과 관련된 대부분의 산  

- volatile acidity : 휘발성 산도: 와인에 함유된 아세트산의 양. 너무 높으면 불쾌한 식초 맛이 날 수 있음  
신맛. 
- citric acid : 구연산: 소량으로 발견되며, 와인에 풍미를 더할 수 있음
- residual sugar : 잔여 당분: 발효가 멈춘 후 남은 설탕의 양으로 1g/L 미만의 와인은 드물며 45g/L 이상의 와인은 단맛으로 간주함  
단맛.
- chlorides : 염소화물: 와인의 염분량
- free sulfur dioxide : 유리 이산화황: 미생물의 성장과 와인의 산화를 방지함
- total sulfur dioxide : 총 이산화황: 저농도에서는 대부분 맛이 나지 않으나 50ppm 이상의 농도에서 맛에서 뚜렷하게 나타남  
화이트 > 레드. 
- density : 밀도: 알코올 및 당 함량에 따라 변함  
바디감. 묵직함  
데이터에서는 알코올, 잔여당분과 상관관계가 보이지는 않음
- pH : 산성 또는 염기성 정도. 0(매우 산성) ~ 14(매우 염기성). 대부분의 와인은 pH 3-4 사이임  
산도와 상관관계가 안보임.  
!퀄리티와 관계가 없어보임
- sulphates : 황산염: 이산화황 농도에 기여할 수 있는 와인 첨가제. 항균 및 항산화제로 작용  
레드와인에서 퀄리티와 상관관계를 보임
- alcohol : 와인의 알코올 함량 백분율  
퀄리티와 낮은 상관관계를 보인다.
- type : 와인에 사용된 포도의 종류. Red(적포도주), White(백포도주)로 나뉨
- quality : 맛으로 평가된 와인의 품질

칼럼이 많다보니 feature selection이 필수적. 이 부분에 많은 시간을 들일 필요가 있다.

맛 -> 바디감, 탄닌, 산도(산미), 당도, 알코올
- 바디감  
묵직함. 액체가 얼마나 끈적거리는가. 알코올을 많이 함유할수록 더 끈적거림
- 탄닌  
떫은 맛. 쌉사름한 맛.  
- 산도  
식초와 같이 톡 쏘는 시큼한 맛을 의미.  
- 당도  
단맛, 달콤한 맛. 맛의 풍미와 균형감을 결정하는데에 큰 역할을 하는 것.
- 알코올  
도수함량, 도수. 라이트바디감(12.5도) / 미디엄 바디감(12.5~13.5도) / 풀바디감(13도 이상)  
낮은 상관관계가 보인다.

전체적인 밸런스가 잘 맞는게 중요하다

어떤 와인이 좋은 와인일까?  이 특성들을 알아내고 주어진 데이터로 만들자  
너무 개인취향을 타는 문제 아닌가? 절대적으로 좋은 와인이라는게 있을까?

좋은 레드 와인은 신맛·떫은맛·쓴맛이 조화를 이루어 어느 하나의 맛이 튀지 않는다.  
일반적으로 화이트 와인에서는 떫은맛·쓴맛이 거의 느껴지지 않는다.  
화이트 와인에서 중요하게 고려되는 것은 신맛과 단맛이다. 신맛과 단맛이 조화를 이루어야 좋은 화이트 와인이다.

색상, 향, 맛, 농밀도, 여운, 

화이트 와인일 때랑 레드 와인일 때랑 각 특성들이 미치는 중요도가 다르지않을까?  
화이트랑, 레드 나눠서 중요 요인 만들어야겠다.

와인의 맛을 결정하는 주된 요인은 포도의 종이다. 기초로 쓰는 포도의 종이 와인의 맛을 결정하는 중요한 요인이다.

pH는 맛에 영향을 안미치는 것으로 보인다.

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split



listdir = os.listdir('data')
print(listdir)

train = pd.read_csv('data/' + listdir[2])
test = pd.read_csv('data/' + listdir[1])
sub = pd.read_csv('data/' + listdir[0])
```

휘발성 산도에서 차이를 보임. 휘발성산도는 신맛이 나게만드는 것인데 

```python
a = list(train.columns)
a.remove('id');a.remove('type')

_, axes = plt.subplots(12, 1, figsize=(12, 40))

for i, col in enumerate(a):
    sns.boxplot(data=train, x='type', y=col, ax=axes[i], showfliers=False)
```

```python
a = list(train.columns)
a.remove('id');a.remove('type')

_, axes = plt.subplots(12, 1, figsize=(12, 40))

for i, col in enumerate(a):
    sns.boxplot(data=train, x='type', y=col, ax=axes[i], showfliers=False, hue='quality')
```

## pH -> acidity 관계파악

```python
# acid, fixed acidity, volatile acidity 관계파악. 
_, axes = plt.subplots(1, 2, figsize=(12, 8))
sns.scatterplot(data=train, x='pH', y='fixed acidity', ax=axes[0])
axes[0].set_title('pH, fixed acidity', color='white')
sns.scatterplot(data=train, x='pH', y='volatile acidity', ax=axes[1])
axes[1].set_title('pH, volatile acidity', color='white')

plt.tight_layout()
```

## density 특성파악

```python
# density 특성파악. 
_, axes = plt.subplots(1, 2, figsize=(12, 8))
sns.scatterplot(data=train, x='alcohol', y='density', ax=axes[0])
axes[0].set_title('alcohol, density')
sns.scatterplot(data=train, x='residual sugar', y='density', ax=axes[1])
axes[1].set_title('residual sugar, density')

plt.tight_layout()
```

## quality 분포확인

## quality별 특성들의 평균값

```python
means = train.groupby('quality').mean().drop(['id'], axis=1)
means
```

```python
fig, axes = plt.subplots(4, 3, figsize=(25, 15))

fig.suptitle('mean values per quality', fontsize= 40)
for ax, col in zip(axes.flat, means.columns):
    ax.plot([4,5,6,7,8], means[col])
    ax.scatter([4,5,6,7,8], means[col])
    ax.set_title(col, fontsize=20, color='white')
    ax.tick_params(labelsize=(20))

plt.setp(axes, xticks=[4, 5, 6, 7, 8])
plt.tight_layout()
plt.show()
```

# modeling

```python
xgb_clf = XGBClassifier(n_estimators=1000, learning_rate=0.05, colsample_bytree=0.5, subsample=0.8)
```

```python
X_features = train.drop([['id', 'pH']], axis=1)
X_features = pd.get_dummies(X_features, drop_first=True)

y_target = train[['quality']]
```

```python
test_X = test.drop([['id', 'pH']], axis=1)
test_X = pd.get_dummies(test_X, drop_first=True)
```

```python
xgb_clf.fit(X_features, y_target, eval_metric='auc')
```

```python
test_pred = xgb_clf.predict(test_X)
```

```python
sub['quality'] = test_pred
```

```python
sub.to_csv('sub/sub_select_column.csv', index=False)
```
