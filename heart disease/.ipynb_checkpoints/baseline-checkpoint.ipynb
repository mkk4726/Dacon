# dataload

id: 데이터 고유 id  
age: 나이  
sex: 성별 (여자 = 0, 남자 = 1)   
cp: 가슴 통증(chest pain) 종류   
- 0 : asymptomatic 무증상
- 1 : atypical angina 일반적이지 않은 협심증
- 2 : non-anginal pain 협심증이 아닌 통증
- 3 : typical angina 일반적인 협심증  

trestbps: (resting blood pressure) 휴식 중 혈압(mmHg)  
chol: (serum cholestoral) 혈중 콜레스테롤 (mg/dl)  
fbs: (fasting blood sugar) 공복 중 혈당 (120 mg/dl 이하일 시 = 0, 초과일 시 = 1)    
restecg: (resting electrocardiographic) 휴식 중 심전도 결과   
- 0: showing probable or definite left ventricular hypertrophy by Estes' criteria
- 1: 정상
- 2: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)

thalach: (maximum heart rate achieved) 최대 심박수  
exang: (exercise induced angina) 활동으로 인한  협심증 여부 (없음 = 0, 있음 = 1)  
oldpeak: (ST depression induced by exercise relative to rest) 휴식 대비 운동으로 인한 ST 하강  
slope: (the slope of the peak exercise ST segment) 활동 ST 분절 피크의 기울기  
- 0: downsloping 하강
- 1: flat 평탄
- 2: upsloping 상승

ca: number of major vessels colored by flouroscopy 형광 투시로 확인된 주요 혈관 수 (0~3 개)
(Null 값은 숫자 4로 인코딩됨)  
thal: thalassemia 지중해빈혈 여부  
- 0 = Null 
- 1 = normal 정상
- 2 = fixed defect 고정 결함
- 3 = reversable defect 가역 결함

target: 심장 질환 진단 여부  
- 0: < 50% diameter narrowing  
- 1: > 50% diameter narrowing

# Modeling

```python
object_cols = ['ca', 'slope', 'restecg', 'cp', 'thal']
another_cols = ['age', 'sex', 'trestbps', 'chol', 'fbs', 'thalach', 'exang', 'oldpeak',  'target']
```

```python
using_cols = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
       'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
# using_cols = ['age', 'sex', 'cp', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'fbs', 'target']
```

```python
def preprocessing(df):
    temp1 = df[object_cols].copy()
    temp2 = df[another_cols].copy()
    
    temp1 = pd.get_dummies(temp1, columns=object_cols)
    
    scaler = StandardScaler()
    scaled = scaler.fit_transform(temp2)
    temp2 = pd.DataFrame(data=scaled, columns=temp2.columns)
    temp1.reset_index(drop=True, inplace=True)
    temp2.reset_index(drop=True, inplace=True)
    df = pd.concat([temp1, temp2], axis=1)
#     df = df[using_cols]
    
    return df
```

```python
def get_train_test(for_train=True, drop_cols=None):
    train, test, sub = load_dataset()
    
    target = train['target']
    total = pd.concat([train, test])
    total.drop(['id'], axis=1, inplace=True)
#     total = preprocessing(total)
    if drop_cols != None:
        total.drop(drop_cols, axis=1, inplace=True)
    train = total[~total['target'].isnull()]
    test = total[total['target'].isnull()]
    
    ftr_x = train.drop(['target'], axis=1)
    
    
    train_x, valid_x, train_y, valid_y = train_test_split(ftr_x, target, 
                                                          test_size=0.2, random_state=100, stratify=target)
    if for_train:
        return train_x, valid_x, train_y, valid_y
    else:
        test.drop(['target'], axis=1, inplace=True)
        return ftr_x, target, test
```

```python
def make_sub(model, drop_cols=None):
    _, _, sub = load_dataset()
    ftr_x, target, test = get_train_test(False, drop_cols)
    
    model.fit(ftr_x, target)
    pred = model.predict(test)
    sub['target'] = pred
    return sub
```

```python
def fit_test(model, drop_cols=None):
    train_x, valid_x, train_y, valid_y = get_train_test()
    model.fit(train_x, train_y)
#     pred = model.predict(valid_x)
#     print(f1_score(valid_y, pred))
    ftr_x, target, test = get_train_test(False, drop_cols)
    scores = cross_val_score(model, ftr_x, target, cv=30, scoring=scorer)
    print('{:.4f}'.format(np.mean(scores)))
```

```python
def scorer(model, X, y):
    pred = model.predict(X)
    return f1_score(y, pred)
```

```python
bayesian_params = {
    'C':(0.1, 1)
}

def svc_eval(C):
    C = max(1e-6, min(C, 1))
    svc_model = SVC(kernel='linear', C=C)
    svc_model.fit(train_x, train_y)
    pred = svc_model.predict(valid_x)
    score = f1_score(valid_y, pred)
    
    return score
```

```python
# BayesianOptimization객체를 수행할 함수와 search할 parameter 범위를 설정하여 생성. 
svcBO = BayesianOptimization(svc_eval, bayesian_params , random_state=0)
# 함수 반환값이 최대가 되는 입력값 유추를 위한 iteration 수행. 
svcBO.maximize(init_points=5, n_iter=25)  
```

```python
results=[]
for result in svcBO.res:
    results.append(result['target'])
    max_ind = np.argmax(results)
print(svcBO.res[max_ind])
```

```python
def get_top_bottom_coef(model):
    train_x, valid_x, train_y, valid_y = get_train_test()
    coef = pd.Series(model.coef_[0], index=train_x.columns)
    coef.sort_values(inplace=True, ascending=False)
    return coef
```

```python
def plot_coef(model):
    coef =  get_top_bottom_coef(model)
    plt.figure(figsize=(12, 6))
    a = sns.barplot(x=coef.values, y=coef.index)
    a.tick_params(axis='y', colors='white')
```

```python
bayesian_params = {
    'C':(0, 0.7)
}

def lg_eval(C):
    C = max(1e-6, min(C, 0.7))
    lg_reg = LogisticRegression(C=C)
    scores = cross_val_score(lg_reg, ftr_x, target, cv=30, scoring=scorer)
    
    return np.mean(scores)
```

```python
bayesian_params = {
    'C':(0, 1),
    'l1_ratio': (0, 1)
}

def lg_eval(C, l1_ratio):
    C = max(1e-6, min(C, 0.9))
    l1_ratio = max(1e-6, min(l1_ratio, 0.9))
    lg_reg = LogisticRegression(penalty='elasticnet', C=C, l1_ratio=l1_ratio, solver='saga')
    scores = cross_val_score(lg_reg, ftr_x, target, cv=30, scoring=scorer)
    return np.mean(scores)
```
