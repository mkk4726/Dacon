# 손 글씨 (숫자) 분류 경진대회 EDA

이번 대회는 Computer Vision 분야에서 가장 유명한 데이터인 손글씨 데이터를 사용하는 대회입니다.

28 x 28 pixel 크기의 손글씨 이미지가 어떤 숫자를 담고 있는지 분류하는 것이 목표입니다.

시작해볼까요?

## 데이터 불러오기

이번 대회의 데이터는 5,000 개의 png 파일과 각 이미지의 정답 label이 담긴 csv 파일로 이루어져 있습니다.

jpg, png와 같은 이미지 파일을 불러오기 위해서 `PIL` 라이브러리를 사용합니다.

```python
import os

datapath = 'data/'
listdir = os.listdir(datapath)
print(listdir)
```

```python
import pandas as pd

train_data = pd.read_csv('data/train/train_data.csv')
train_data.rename(columns={'filen_name':'file_name'}, inplace=True)
train_data
```

```python
import numpy as np
import pandas as pd
from PIL import Image # 이미지를 불러오기 위한 라이브러리
import matplotlib.pyplot as plt # 그림을 표시하기 위한 라이브러리
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

train_data = pd.read_csv('dataset/train/train_data.csv')
train_data
```

데이터셋을 살펴보니 `file_name`에서는 이미지의 파일명, `label`은 그 이미지가 어떤 숫자를 담고 있는지 알 수 있네요.

먼저 숫자 손글씨 이미지란 대체 어떤 이미지인지 살펴볼까요?

```python
datapath = 'data/train/train_data.csv'
```

```python
from PIL import Image

train_file_name = train_data['file_name']
train_label = train_data['label']

# image 파일을 불러와서 list에 담아줍니다.
train_image = []
for file in train_file_name:
    train_image.append(Image.open("data/train/"+file))
```

```python
import matplotlib.pyplot as plt

# 랜덤으로 9개를 뽑아서 이미지를 표시해봅시다.
random_images = np.random.permutation(train_image)[:9] 

plt.style.use('ggplot')
fig, ax = plt.subplots(3,3, figsize=(15,15))
fig.suptitle("Random Digits", fontsize=40)
for i in range(1,10):
    plt.subplot(3,3,i)
    plt.imshow(random_images[i-1])
plt.show()
```

음 이렇게 생긴 숫자들이군요!

화질이 조금 좋지 않지만 그래도 육안으로 봤을 때 어떤 숫자인지는 알 수 있을 것 같습니다.

그러면 각 숫자들은 몇개씩 주어졌을까요?

```python
from collections import Counter

label_counter = Counter(train_label)

plt.figure(figsize=(10,10))
plt.title("Number Counts", fontsize=25)
plt.bar(x=list(label_counter.keys()), height =list(label_counter.values()))
plt.xticks(list(range(10)))
plt.show()

for i in range(10):
    print(f"숫자 {i} 의 개수는 {label_counter[i]}개 입니다.")
```

와! 이미지가 각 숫자별로 모두 동일하게 500개씩 있네요!

## 데이터 살펴보기

이제 이미지를 자세히 살펴봅시다.

이미지를 "분석"하기 위해서는 먼저 이미지를 숫자로 표현해야 합니다.

그림과 그림은 더하거나 뺄 수 없지만, 숫자는 그게 가능하니까요!

먼저 이미지 하나를 골라서 분석해봅시다.

```python
image, label = train_image[77], train_label[77]

plt.figure(figsize=(10,10))
plt.imshow(image)
plt.title("Label : "+str(label), fontsize=25)
plt.show()
```

```python
# 이 이미지의 간단한 통계 정보를 살펴봅시다.
image = np.array(image)

print("이미지의 사이즈 : ", image.shape)
print("이미지의 최소, 최대 값 : ", image.min(), image.max())
print("이미지의 유의미한 pixel 개수 : ", (image > 0).sum())
```

이미지는 가로로 28 pixel, 세로로 28 pixel인 이미지라는 것을 알 수 있습니다.  

각 픽셀의 값은 0 에서 254의 값을 갖고 있고,  

위 이미지의 경우 129 개의 픽셀로 숫자 9를 표현했네요.

이번엔 이미지들 전체를 살펴봅시다.

```python
image2number.shape
```

```python
# 먼저 이미지 전체를 np.array 형태로 변환해줍니다.
# 또한 분석의 편의성을 위해서 flatten() 메소드를 사용해서 28 x 28 이미지를 28*28=748 array로 쭈욱 펴줄게요.
image2number = np.array([np.array(image).flatten() for image in train_image])

print("이미지의 최소, 최대 값 : ", image2number.min(), image2number.max())
print("이미지의 유의미한 pixel 개수 평균 : ", (image2number > 0).sum()/len(image2number))
```

하나의 이미지만 봤을 때는 이미지의 최대 값이 254 였는데, 전체 이미지를 살펴보니 최대값이 255라는 것을 알 수 있었네요.

또, 평균적으로 151개의 픽셀을 사용해서 숫자를 표현했다는 것을 알 수 있었습니다.

이번에는 각 숫자들별로 어떤 특징을 가지고 있는지 살펴봐요.

먼저 숫자들별로 이미지를 분류해줍니다.

```python
# 0 부터 9 까지의 key 값을 가지고 value로 이미지를 가진 dictionary를 구성합니다.
image_dict = {}

for i in range(10):
    idx = train_label[train_label==i].index
    image_dict[i] = image2number[idx]
```

이제 이 데이터로 사람들이 "평균적으로" 어떻게 숫자를 쓰는지 한번 알아봅시다!

각각 숫자 이미지들을 모두 더해준 다음 시각화하면 됩니다.

```python
plt.figure(figsize=(15,8))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.grid(False)
    plt.title("Label : "+str(i), fontsize=25)
    plt.imshow(image_dict[i].sum(axis=0).reshape((28,28)))
plt.show()
```

평균적으로 보니 생각보다 깔끔한 숫자들이 나왔네요. 

나중에 숫자를 예측하는 모델을 만들 때 이런 정보를 사용해도 좋겠죠?

## 좀 더 복잡한 시각화

각 숫자별로 이미지들이 어떤 공통된 특징을 갖는지 알아봤습니다.

이제는 데이터셋을 전체적으로 봤을 때, 각 숫자들만의 특성이 있는지 한번 알아보도록 합시다.

차원 축소(Dimensionality Reduction) 방법 중 하나인 T-SNE 를 사용할건데요,

이론적으로 어려운 내용이고 실제 분석에는 사용할 일이 없을테니 여기서는 어떤 그림이 나오는지만 보도록 합시다.

```python
# sklearn 라이브러리를 사용합니다.
from sklearn.manifold import TSNE

# 3~5분 정도 걸려요!
image_tsne = TSNE(n_components=2).fit_transform(image2number)
```

```python
# 예쁜 그림을 그리기 위해 seaborn 라이브러리를 사용합니다.
import seaborn as sns

plt.figure(figsize=(16,10))
plt.title("Digits T-SNE visualization", fontsize=25)
sns.scatterplot(
    x=image_tsne[:,0], y=image_tsne[:,1],
    hue=train_label,
    palette=sns.color_palette("hls", 10),
    legend="full"
)
plt.show()
```

오래 기다리셨습니다! 모든 이미지들을 한번에 보기 쉽도록 그들의 특성만 뽑아서 그래프를 그렸습니다.

그림을 보시면 숫자 0, 1, 2, 6, 7 은 자기들끼리 잘 뭉쳐있는 모습을 보이는 반면,

숫자 3, 4, 5, 8, 9 는 서로 섞여 있어서 쉽게 구분하기 힘들 것 같다는 생각이 드네요.
